```python
import pandas as pd

# 用pd.read_csv()函数读取CSV文件
df = pd.read_csv('listings-2023-drop_cols.csv')

# 打印前几行数据以进行查看
print(df.head())
```

```python
money = ['price']
for m in money:
    print(f"Converting {m}")
    df[m] = df[m].str.replace('$','', regex=False).str.replace(',','', regex=False).astype('float')
```

```python
value_ave = df.groupby('neighbourhood_cleansed')['review_scores_value'].mean().reset_index()
```

```python
price_ave = df.groupby('neighbourhood_cleansed')['price'].mean().reset_index()
```

```python
merged_df = pd.merge(value_ave, price_ave, on='neighbourhood_cleansed', how='inner')
```
